# **Gym Member Behavior and Engagement Patterns**


A gym wants to improve member engagement, retention, and personalized fitness offerings.


By analysing gym membership data, we aim to:
* Understand member demographics
* Analyse gym usage patterns
* Identify popular workout habits & group classes

# **Domain**

# **Objective**

#### The primary objective of this dataset is to analyze gym member behavior and demographics to improve service delivery and member engagement. Specifically, it aims to:



1. Understand patterns in gym usage (frequency, duration, and timing).

2. Assess the popularity of different membership tiers (Standard vs. Premium).

3. Evaluate the uptake and preferences for specialized services like Group Lessons and Personal Training.

4. Segment members based on age, gender, and habits to potentially offer personalized fitness plans.

#### The expected outcomes of analyzing this data include:

* Operational Optimization: Identifying peak check-in and check-out times to manage staff and equipment availability.

* Membership Insights: Determining which membership types provide better retention or higher engagement.

* Program Development: Identifying popular group lessons (e.g., Kickboxen, XCore, Yoga) to schedule them more effectively.

* Targeted Marketing: Creating profiles for members who are likely to hire a Personal Trainer or upgrade their membership.

## Stage 2 – Data Cleaning and Pre-processing

Handle missing values (impute or drop)

Handle duplicates

Treat outliers if required

Check skewness and apply transformations

Convert data types if needed

Feature transformations (date parts, derived fields if required for analysis)

```python
# Handling Missing Values

print(df.isnull().sum())


# Fill missing group lessons with "None"
df['Fav Group Lesson'] = df['Fav Group Lesson'].fillna("None")


# Fill missing trainer names
df['Personal Trainer Name'] = df['Personal Trainer Name'].fillna("Not Assigned")
```

```python
df.info()
```

```python
#Handling Duplicate Records

df.duplicated().sum()
```

```python
#If duplicates exist, they are removed

df = df.drop_duplicates()
```

```python
#Identify Outliers (Numerical Columns)

df[['Age', 'Visit Per Week', 'Avg time in gym']].describe()
```

```python
#Outlier Check Using IQR Method (Interquartile Range)

Q1 = df['Avg time in gym'].quantile(0.25)
Q3 = df['Avg time in gym'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df = df[(df['Avg time in gym'] >= lower_bound) &
        (df['Avg time in gym'] <= upper_bound)]
```

```python
#Check Skewness

df['Avg time in gym'].skew()
```

```python
#Apply Transformation (If Skewed)
import numpy as np

df['Avg time in gym_log'] = np.log1p(df['Avg time in gym'])
```

```python
#Convert Date Column

df['DOB'] = pd.to_datetime(df['DOB'])
```

```python
#Convert Time Columns
df['Avd Check -in time'] = pd.to_datetime(df['Avd Check -in time'])
df['Avg Check out time'] = pd.to_datetime(df['Avg Check out time'])
```

```python
#Extract Date Features

df['Birth_Year'] = df['DOB'].dt.year
df['Birth_Month'] = df['DOB'].dt.month
```

```python
#Age Group Creation

bins = [0, 20, 30, 40, 50, 100]
labels = ['Teen', 'Young Adult', 'Adult', 'Mid Age', 'Senior']

df['Age Group'] = pd.cut(df['Age'], bins=bins, labels=labels)
```

```python
#Workout Intensity Level

df['Workout Intensity'] = df['Visit Per Week'].apply(
    lambda x: 'Low' if x <= 2 else 'Medium' if x <= 4 else 'High'
)
```

```python
#Final Dataset Check

df.info()
df.head()
```

### Chart 1: Gender Distribution (Count Plot)

```python
import matplotlib.pyplot as plt

gender_counts = df['Gender'].value_counts()

bars = plt.bar(gender_counts.index, gender_counts.values)

# Add value labels on top of bars
for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width()/2,
        height,
        str(height),
        ha='center',
        va='bottom'
    )

plt.title("Gender Distribution of Gym Members")
plt.xlabel("Gender")
plt.ylabel("Number of Members")
plt.show()
```

## Interpretation

The chart shows a relatively balanced gender distribution.



### **Insights**

* The gym caters to both genders equally, indicating that fitness programs and marketing strategies should remain gender-inclusive.

# Chart 2: Age Distribution (Histogram)

```python

# Create histogram with fewer bins to increase gap
counts, bins, patches = plt.hist(
    df['Age'],
    bins=10,          # fewer bins = wider bars + more gap
    rwidth=0.8        # controls gap between bars
)

# Add value labels on each bar
for count, patch in zip(counts, patches):
    if count > 0:
        plt.text(
            patch.get_x() + patch.get_width() / 2,
            count,
            int(count),
            ha='center',
            va='bottom'
        )

plt.title("Age Distribution of Gym Members")
plt.xlabel("Age")
plt.ylabel("Number of Members")
plt.show()

```

## Interpretation of your above chart

Most members fall between 20 and 40 years.

# Business Insight:

This age group prefers high-energy workouts such as HIIT, strength training, and group classes, which should be prioritized.

#Chart 3: Average Time Spent in Gym (Boxplot)

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 4))

plt.boxplot(
    df['Avg time in gym'],
    vert=True,
    patch_artist=True,
    boxprops=dict(facecolor='lightblue'),
    medianprops=dict(color='red', linewidth=2),
    whiskerprops=dict(color='black'),
    capprops=dict(color='black'),
    flierprops=dict(marker='o', markerfacecolor='orange', markersize=5)
)

plt.title("Average Time Spent in Gym (in Minutes)", fontsize=12)
plt.ylabel("Minutes", fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.6)

plt.show()

```

## Interpretation of your above chart

Most members spend 90–150 minutes per session, with few extreme values.

# Business Insight:
Session durations indicate serious fitness engagement, making members suitable candidates for premium plans and personal training.

#Chart 4: Age vs Visits Per Week (Scatter Plot)

```python
import numpy as np

plt.figure(figsize=(7, 5))

plt.scatter(
    df['Age'],
    df['Visit Per Week'],
    alpha=0.6,          # transparency for overlap
    s=60,               # marker size
    edgecolors='black'
)

# Add trend line
z = np.polyfit(df['Age'], df['Visit Per Week'], 1)
p = np.poly1d(z)
plt.plot(df['Age'], p(df['Age']), linestyle='--', linewidth=2, label='Trend Line')

plt.title("Relationship Between Age and Gym Visits Per Week", fontsize=12)
plt.xlabel("Age", fontsize=10)
plt.ylabel("Visits Per Week", fontsize=10)

plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.show()
```

## Interpretation of your above chart

Younger members tend to visit the gym more frequently.

#Business Insight:

Retention strategies for older members (e.g., wellness programs, flexible workouts) could increase their visit frequency.

#Chart 5: Membership Type vs Visit Frequency (Bar Plot)

```python
avg_visits = df.groupby('MemberShip Type')['Visit Per Week'].mean()

plt.figure(figsize=(7, 5))

bars = plt.bar(
    avg_visits.index,
    avg_visits.values,
    edgecolor='black'
)

# Add value labels on bars
for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width() / 2,
        height,
        f"{height:.2f}",
        ha='center',
        va='bottom',
        fontsize=10
    )

plt.title("Average Visits Per Week by Membership Type", fontsize=13)
plt.xlabel("Membership Type", fontsize=11)
plt.ylabel("Average Visits per Week", fontsize=11)

plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

## Interpretation:

Premium members visit the gym more often than basic members.

## Business Insight:
Encouraging upgrades to premium plans could improve engagement and revenue.

#Chart 6: Correlation Heatmap (Numerical Variables)

```python
import numpy as np
import matplotlib.pyplot as plt

# Correlation matrix
corr = df[['Age', 'Visit Per Week', 'Avg time in gym']].corr()

plt.figure(figsize=(6, 5))

# Heatmap with color variation
im = plt.imshow(corr, cmap='coolwarm', vmin=-1, vmax=1)

# Axis labels
plt.xticks(range(len(corr.columns)), corr.columns, rotation=45)
plt.yticks(range(len(corr.columns)), corr.columns)

# Add correlation values inside cells
for i in range(len(corr.columns)):
    for j in range(len(corr.columns)):
        plt.text(
            j, i,
            f"{corr.iloc[i, j]:.2f}",
            ha="center",
            va="center",
            color="black",
            fontsize=10
        )

plt.title("Correlation Heatmap of Gym Variables", fontsize=13)
plt.colorbar(im)
plt.tight_layout()
plt.show()

```

## Interpretation:

Visit frequency has a stronger relationship with time spent in the gym than age.

## Business Insight:
Frequent visitors are more engaged regardless of age, making them ideal targets for loyalty programs.

# Chart 7: Visits Per Week by Age Group and Gender (Grouped Analysis)

```python
df.groupby(['Age Group', 'Gender'])['Visit Per Week'].mean().unstack().plot(kind='bar')
plt.title("Average Visits Per Week by Age Group and Gender")
plt.xlabel("Age Group")
plt.ylabel("Average Visits")
plt.show()
```

# Interpretation:

Young adults of both genders show higher gym engagement.

# Business Insight:
Targeted marketing campaigns for young adults can maximize gym utilization.

# Chart 8: Personal Training Preference by Age Group

```python
df.groupby('Age Group')['Personal Training'].value_counts().unstack().plot(kind='bar')
plt.title("Personal Training Preference by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Number of Members")
plt.show()

```

## Interpretation:
Older members show a higher preference for personal training.

## Business Insight:
Personalized coaching packages should be marketed more aggressively to older age groups.

#Chart 9: Pivot Table – Workout Intensity vs Group Lessons

```python
import matplotlib.pyplot as plt
import numpy as np

pivot = pd.pivot_table(
    df,
    values='Visit Per Week',
    index='Workout Intensity',
    columns='Attend Group Lesson',
    aggfunc='mean'
)

plt.figure(figsize=(8, 5))

bars = pivot.plot(
    kind='bar',
    edgecolor='black'
)

# Add value labels
for container in bars.containers:
    bars.bar_label(container, fmt='%.2f', padding=3)

plt.title("Workout Intensity vs Group Lesson Participation", fontsize=13)
plt.xlabel("Workout Intensity Level", fontsize=11)
plt.ylabel("Average Visits Per Week", fontsize=11)

plt.legend(title="Attend Group Lesson")
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

#Interpretation:
Members with high workout intensity often attend group lessons.
#Business Insight:
Group classes play a major role in retaining highly active members.
